```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def gabor(XX, YY, theta, a, b, F):
    x_r = XX * np.cos(theta) + YY * np.sin(theta)
    y_r = -XX * np.sin(theta) + YY * np.cos(theta)
    gaussian = np.exp(-np.pi*(a**2*x_r**2+b**2*y_r**2))
    complex_grating = np.exp(1j*2*np.pi*F*(XX*np.cos(theta)+YY*np.sin(theta)))
    return gaussian * complex_grating.real
```

```python
gabor_width = 0.15
gabor_height = 0.1
gabor_freq = 0.15
gabor_angle = np.pi/2

XX, YY = np.ogrid[-20:20:1000j,-20:20:1000j]
W = gabor(XX, YY, gabor_angle, gabor_width, gabor_height, gabor_freq)
```

```python
plt.imshow(W)
plt.gray()
plt.colorbar()
```

```python
def bar(XX, YY, x0, y0, length, width, angle):
    c = np.cos(angle)
    s = np.sin(angle)
    dx = (XX - x0)
    dy = (YY - y0)
    return  (((np.abs((dy * s) - (dx * c)) < length / 2))) & ((np.abs((dx * s) + (dy * c)) < width / 2))
```

```python
bar_length = 10.
bar_width = 2.
bar_angle = 0.
b = bar(XX, YY, 0, 0, bar_length, bar_width, bar_angle)
plt.imshow(b)
plt.gray()
```

```python
def simple_cell_model(receptive_field):
    def _response(stim):
        r = np.sum(stim * receptive_field)
        return r
    return _response
```

```python
simple_cell = simple_cell_model(W) 
```

```python
def position_tuning(response_func):
    xcoords = np.linspace(-10, 10, 100)
    responses = []
    for x0 in xcoords:
        b = bar(XX, YY, 0, x0, 10., 1.8, np.pi)
        r = response_func(b)
        responses.append(r)
    responses = np.array(responses)
    return xcoords, responses
```

```python
xcoords, responses = position_tuning(simple_cell)
plt.plot(xcoords, responses)
```

```python
plt.figure(figsize=(9, 3))
plt.subplot(121)
plt.imshow(W)
b1 = bar(XX, YY, 0, 0, 10., 2.5, np.pi)
b2 = bar(XX, YY, 0, 2.5, 10., 2.5, np.pi)
plt.contour(b1)
plt.contour(b2)
plt.subplot(122)
r1 = simple_cell(b1)
r2 = simple_cell(b2)
r1and2 = simple_cell(b1 + b2)
plt.bar(range(4), [r1, r2, r1+r2, r1and2])
plt.xticks(np.arange(4) + 0.5, ["Bar 1", "Bar 2", "sum of\n responses", "Bar 1 + 2"])
plt.ylabel('response')
```

```python
def orientation_tuning(response_func):
    responses = []
    thetas = np.linspace(-np.pi, np.pi, 100)
    for theta in thetas:
        b = bar(XX, YY, 0, 0, 5., 1., theta)
        r = response_func(b)
        responses.append(r)
    responses = np.array(responses)
    return thetas, responses
```

```python
thetas, responses = orientation_tuning(simple_cell)
plt.plot(thetas, responses)
```

```python
def complex_cell_model(receptive_fields):
    simple_cells = [simple_cell_model(rf) for rf in receptive_fields]
    
    def _response(stim):
        response = 0
        for inp in simple_cells:
            response += inp(stim) ** 2
        return response
    return _response
```
